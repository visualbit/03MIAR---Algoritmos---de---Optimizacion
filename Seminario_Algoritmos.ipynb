Diseña un algoritmo para resolver el problema por fuerza bruta

Respuesta

```
import itertools
import random

"""
    Función que recibe una lista de números y un objetivo y retorna todas las posibles combinaciones
    del operaciones para obtener el objetivo que es enviado por parámetro, o None si no es posible.

    Este algoritmo utiliza la fuerza bruta y obtiene todas las permutaciones de los números enviados 
    en la lista y los operadores, que logren el objetivo

    Args:
        lista (list): lista de números enteros.
        objetivo (int): número entero objetivo.

    Returns:
        str or None: expresión matemática que utiliza todos los números de la lista combinados
        con los operadores para obtener el objetivo, o None si no es posible.
"""

def combinar_cifras_operaciones(lista, objetivo):

    operadores = ['+', '-', '*', '/']
 
    #utilizo la función permutations para crear las combinaciones de los operadores sin repetición
    for op1, op2, op3, op4 in itertools.permutations(operadores, 4):
    #Recorro la lists de números y realizo todas las permutaciones posibles de la lista, 
    #con una longitud que no supere la longitud de los operadores +1, 
    #para que no se repitan operadores
        for x in itertools.permutations(lista, len(operadores)+1):      
            #Realizo las operaciones necesarias para llegar al objetivo (Fuerza bruta)
            exp = f"{x[0]}{op1}{x[1]}{op2}{x[2]}{op3}{x[3]}{op4}{x[4]}"
            try:
              #Evalúo la expresión resultante con el objetivo propuesto, si es igual, lo imprimo
                if eval(exp) == objetivo:
                    print(exp, "=", objetivo)
            except ZeroDivisionError:
                pass
                


numero = 22
lista_numeros = [1,2,3,4,5,6,7,8,9]
resultado = combinar_cifras_operaciones(lista_numeros, numero)
print(resultado)

```

Calcula la complejidad del algoritmo por fuerza bruta

Respuesta

La complejidad del algoritmo es O(n!) porque utiliza la función permutations de la biblioteca itertools, que genera todas las posibles permutaciones de la lista de números y los operadores, lo que implica una complejidad factorial en el peor caso.

Además, para cada permutación, se realiza una evaluación de la expresión y una comparación con el objetivo, lo que implica una complejidad adicional de O(n). Por lo tanto, la complejidad total del algoritmo es O(n! * n).

(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Respuesta

Una posible solución para mejorar el algoritmo por fuerza bruta es utilizar la técnica de búsqueda exhaustiva con recursión y poda. En este algoritmo, se evalúa recursivamente todas las posibles combinaciones de números y operadores, y se utiliza la poda para descartar las combinaciones que no pueden alcanzar el objetivo, finalmente se imprime la primer solución encontrada. Es una técnica que mejora en rendimiento porque no tiene en cuenta todas las posibles combinaciones, aunque su complejidad es similar

```
import itertools

class Nodo:
    def __init__(self, cifras, operadores, expresion):
        self.cifras = cifras
        self.operadores = operadores
        self.expresion = expresion

    def obtener_resultado(self):
        try:
            return eval(self.expresion)
        except:
            return None

    def es_meta(self, meta):
        resultado = self.obtener_resultado()
        return resultado is not None and resultado == meta

    def generar_hijos(self):
        hijos = []
        # crear todas las combinaciones posibles de 5 cifras y 4 operadores
        for cifras_combinadas in itertools.permutations(self.cifras, 5):
            for operadores_combinados in itertools.permutations('+-*/', 4):
                # saltar las combinaciones que utilicen un número o un operador más de una vez
                if len(set(cifras_combinadas)) != 5 or len(set(operadores_combinados)) != 4:
                    continue
                # construir la expresión y crear un nuevo hijo
                expresion = f'{cifras_combinadas[0]} {operadores_combinados[0]} {cifras_combinadas[1]} {operadores_combinados[1]} {cifras_combinadas[2]} {operadores_combinados[2]} {cifras_combinadas[3]} {operadores_combinados[3]} {cifras_combinadas[4]}'
                hijos.append(Nodo([], [], expresion))
        return hijos

def ramificacion_y_poda(nodo_inicial, meta):
    frontera = [nodo_inicial]
    while len(frontera) > 0:
        nodo_actual = frontera.pop(0)
        if nodo_actual.es_meta(meta):
            print(meta)
            print("¡Solución encontrada!")
            print(f"Expresión: {nodo_actual.expresion}")
            print(f"Resultado: {nodo_actual.obtener_resultado()}")
            return
        hijos = nodo_actual.generar_hijos()
        hijos = [h for h in hijos if h.obtener_resultado() is not None]
        hijos.sort(key=lambda h: abs(h.obtener_resultado() - meta))
        frontera = hijos + frontera
    print("No se encontró solución.")


nodo_inicial = Nodo([1, 2, 3, 4, 5, 6, 7, 8, 9], [], '')
meta = -69
ramificacion_y_poda(nodo_inicial, meta)
```

Respuesta

La complejidad de este algoritmo depende de varios factores, como el tamaño del conjunto de cifras y el valor de la meta. Sin embargo, en términos generales, podemos considerar que su complejidad es exponencial en el peor caso, ya que tiene que explorar todas las combinaciones posibles de cifras y operadores para encontrar una solución. Además, el algoritmo utiliza una estrategia de búsqueda en anchura, lo que significa que puede requerir una gran cantidad de memoria si la profundidad de la búsqueda es grande. En resumen, la complejidad de este algoritmo es alta y puede ser prohibitiva para entradas grandes. Sin embargo su complejidad es menor que la de la fuerza bruta, ya que no se exploran todas las combinaciones, gracias a la poda
